# DATA CLEANING

# Validity

```
# Deleting unnecesarry data
df_merged.drop('Country', axis=1, inplace=True)
```

```
df_merged.drop('Status', axis=1, inplace=True)
```

# Accuracy

```
# changing the format of Disbursement Date and End Date
df_merged['Disbursement Date'] = pd.to_datetime(df_merged['Disbursement Date'])
df_merged['End Date'] = pd.to_datetime(df_merged['End Date'])
```

```
def check(row):
  if row['Disbursement Date'] < row['End Date']:
    return True
  else:
    return False
```

```
df_merged['verifier'] = df_merged.apply(lambda row:check(row), axis=1)
df_merged.head()
```
